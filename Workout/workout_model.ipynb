```python
wo_data = pd.read_csv("./workout.csv",delimiter=';')
wo_data
```

```python
x = ['Type', 'BodyPart', 'Equipment', 'Level']
y = 'Title'
```

```python
# Initiate Label Encoding
label_encoder = LabelEncoder()
```

```python
# Initiate Label Encoding on y value
y = label_encoder.fit_transform(wo_data[y])
```

```python
y
```

```python
# Initiate One Hot Encoder on X values
encoders = {}
input_data = []
for feature in x:
    encoder = OneHotEncoder(sparse=False)
    encoding = encoder.fit_transform(np.array(wo_data[feature]).reshape(-1, 1))
    encoders[feature] = encoder
    input_data.append(encoding)
```

```python
# Concatenate the One Hot Encoded value
x1 = np.concatenate(input_data, axis=1)
```

```python
x1
```

```python
# Creating Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(len(set(y)), activation='softmax')
])

# Compiling Model
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Fitting the Model
model.fit(x1, y, epochs=10, batch_size=16)
```

```python
# New Input example
new_input = {
    'Type': ['Strength'],
    'BodyPart': ['Abdominals'],
    'Equipment': ['Barbell'],
    'Level': ['Beginner']
}
```

```python
# Check the possible outcome without using the trained model
mask = pd.Series([True] * len(wo_data))

for column, values in new_input.items():
    mask &= wo_data[column].isin(values)
filtered_df = wo_data[mask]

filtered_df
```

```python
# Do One Hot Encoding on the new_input data
new_input_data = []
for feature in x:
    encoder = encoders[feature]
    encoding = encoder.transform(np.array(new_input[feature]).reshape(-1, 1))
    new_input_data.append(encoding)
```

```python
# Concatenate the One Hot Encoded new_input
new_x = np.concatenate(new_input_data, axis=1)

# Make a prediction using the Model
predictions = model.predict(new_x)
```

```python
# Determine The output count
k = 5
top_k_classes = np.argsort(predictions, axis=1)[:, -k:]
```

```python
# Inverse Transform the Value from the predictions back to string
predicted_labels = label_encoder.inverse_transform(top_k_classes[0])
print(predicted_labels)
```
